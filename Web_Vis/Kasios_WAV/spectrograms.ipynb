```python
def make_spectrogram(file):
    in_file = file
    samplingFrequency, signalData = wavfile.read(in_file)
    title = in_file[0:(len(in_file) - 4)] + " Spectrogram"
    out_file = "./Spectrograms/" + in_file[0:len(in_file) - 4] + ".png"
    # Plot the signal read from wav file
    fig, ax = plt.subplots()
    pxx, freq, t, cax = ax.specgram(signalData,Fs=samplingFrequency, mode='magnitude', noverlap=128, NFFT=256, cmap = cmap, vmin=-5)
    plt.xlabel('Time [Sec]')
    plt.ylabel('Hz')
    plt.ylim(0,15000)
    cbar = fig.colorbar(cax)
    cbar.set_label("dB")
    ax.title.set_text(title)
    plt.savefig(out_file)
```

```python
for root, dirs, files in os.walk("."):
    for filename in files:
        if filename.endswith(".wav"):
            make_spectrogram(filename)
        else:
            continue
```
